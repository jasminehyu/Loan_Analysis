```python
#Q6
loan_tree_uwcu=loans.BST()
num_total=0
time_total=0
times=[]
nums=[]
for item in uwcu.list:
    begin=time.time()
    loan_tree_uwcu.add(item.interest_rate,item)
    finish=time.time()
    time_consumed=(finish-begin)*10**3
    time_total+=time_consumed
    times.append(time_total)
    num_total+=1
    nums.append(num_total)

plt.plot(nums,times)
plt.xlabel("BST Adds")
plt.ylabel("Total Elapsed Milliseconds")
```

```python
#Q8
empty_dict={}
for item in uwcu.list:
    for item in item.applicants:
        num_of_race=len(item.race)
        if not str(num_of_race) in empty_dict:
            empty_dict[str(num_of_race)]=0
            
        empty_dict[str(num_of_race)]+=1
        

keys=empty_dict.keys()
values=[empty_dict[item] for item in empty_dict]

plt.bar(keys,values)
plt.xlabel('Num of races indicated in the profile')
plt.ylabel('Num of applicants')
```
